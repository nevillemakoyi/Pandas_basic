<b> Le concept de tri stable </b>

Le "tri stable" fait référence à la préservation de l'ordre relatif des enregistrements égaux par rapport à la clé de tri. 

En d'autres termes, si deux éléments sont considérés comme égaux dans le contexte de la clé de tri (c'est-à-dire qu'ils ont la même valeur pour cette clé), un tri stable garantira que leur ordre relatif original est préservé dans le résultat trié.

```python
sorted_by_class = data.sort_values(by='Pclass')
sorted_by_class.head()
```

Dans pandas, le tri par défaut (quicksort) n'est pas stable, ce qui signifie que l'ordre relatif des éléments égaux peut être modifié. En spécifiant kind='mergesort', vous pouvez obtenir un tri stable (à noter qu'il est légèrement plus gourmant en ressources).

```python
sorted_by_class = data.sort_values(by='Pclass', kind='mergesort')
sorted_by_class.head()
```
